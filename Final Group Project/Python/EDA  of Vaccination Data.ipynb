# Swarm plot

```python
A swarm plot can be drawn on its own,but it is also a good complement to a box or violin plot in cases
where you want to show all observations along with some representatioin of the underlying distribution.
This style of plot isnsometimes called a beeswarm.
```

```python
sns.swarmplot(d["Dose 1"])
plt.show()
```

```python
sns.swarmplot(x="Dose 1",y="Dose 2",data=d)
plt.show()
```

```python
sns.swarmplot(x="Dose 1",y="Dose 2",hue="State",data=d)
plt.show()
```

# Dist plot
Seaborn distplot lets you show a histogram with a line on it.This can be shown in all kinds of variations.
A distplot plot a univariate distribution of observations.


```python
sns.distplot(d.Active)
plt.show()
```

```python
sns.distplot(d.Active,color="green")
plt.show()
```

```python
sns.distplot(d.Active,color="green",hist=False)
plt.show()
```

```python
sns.distplot(d.Active,color="green",kde=False)
plt.show()
```

```python
sns.distplot(d.Active,color="green",bins=10)
plt.show()
```

```python
plt.figure(figsize=(8,3))
sns.distplot(d.Active,
           bins=20,
           kde_kws={"lw":4,"color":"green"},
            hist_kws={"alpha":0.3,"color":"gray"}
            )
plt.xlabel("weight (lbs)",fontsize=16,labelpad=15)
plt.yticks([])
plt.show()
```
