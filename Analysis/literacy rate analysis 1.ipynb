# Data Visualization

###  bar graph 
it shows the literacy rate total 2001 vs category 

```python
plt.bar(goi['lrtotal1'],goi['Category']) 
plt.title('lrtotal1 vs category',fontsize=15,color='red') 
plt.xlabel('lrtotal1',fontsize=15) 
plt.ylabel('Category',fontsize=15) 
plt.grid(True)
plt.show()
```

###  horizontal bargraph
in this horizontal graph we can easily analyse the  literacy rate of 2001 of states and UT in India

```python
plt.figure(figsize=(10,15))
plt.barh(goi['places'],goi['lrtotal1']) 
plt.title('lrtotal1 vs places',fontsize=15,color='red') 
plt.xlabel('places',fontsize=15) 
plt.ylabel('lrtotal1',fontsize=15) 
plt.grid(True)
plt.show()
```

###  count plot
count plot of column category

```python
sns.countplot(y='Category',data=goi,palette='tab10')
plt.show()
```

###  piechart
in this piechart we try to visualise counts of category 

```python
plt.title('category',fontsize=25)
goi['Category'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,explode=(0,0,0.5))
plt.show()
```

###   bar plot
in this we using groupby command to plot this graph . 

```python
goi.groupby('Category')[['lrtotal1','lrtotal11']].sum().plot.bar(color=['blue','red'],figsize=(8,6))
plt.ylabel('lrtotal11/lrtotal1')
plt.show()
```

### horizontal bargraph 

```python
plt.figure(figsize=(12,8))
plt.title('category wise literacy rate in urban area 2001')
sns.barplot(x='lrurban11',y='places',data=goi,hue='Category',palette='tab10')
plt.xlabel('literacy rate of urban',fontsize=15)
plt.show()
```

### place vs literacy rate of 2001 

```python
goi.plot(kind='bar',figsize=(80,40))
plt.tick_params(axis='x',labelsize=33)
plt.tick_params(axis='y',labelsize=33)
plt.xlabel('place',fontsize=50)
plt.ylabel('lrtotal1',fontsize=50)
plt.title('literacy rate of 2001 across places',fontsize=50)
plt.legend(['lrtotal11','lrrural1'],prop={'size':30})
plt.show()
```

###  strip plot
in this strip plot we try to visualize the literacy rate total 2011 with the category and state category have the most percentage literacy rate in india 

```python
sns.set_style('darkgrid')
sns.stripplot(x='Category',y='lrtotal11',data=goi)
plt.show()
```

###  strip plot

###  swarm plot
this plot shows that the literacy rate total2011 of union territories using swarm plot . and we can see that lakshadweep has the highest literacy rate among the unionterritories

```python
ut=goi[goi.Category=='Union Territory']
sns.swarmplot(x=ut.places,y=ut.lrtotal11,color='xkcd:forest green',size=10)
plt.show()
```

###  distplot
in this distplot we visualize the literacy rate total 2011 against the density distribution.

```python
sns.distplot(goi.lrtotal11,color='orange')
plt.show()
```

```python
sns.distplot(goi.lrtotal11,color='blue',bins=37)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.distplot(goi.lrtotal11,
            bins=25,
           kde_kws={'lw':8,'color':'black'},
           hist_kws={'alpha':0.3,'color':'gray'}
           )
plt.xlabel('lrtotal11',fontsize=16,labelpad=15)
plt.yticks([])
plt.show()
```
