# Data Visualisation

### team vice gold medal using bar graphs
it shows that the highest gold medals won by the team is USA.

```python
plt.figure(figsize=(10,20))
plt.bar(tokyo['gold'],tokyo['team']) 
plt.title('teamvice gold medal',fontsize=15,color='red') 
plt.xlabel('gold',fontsize=15) 
plt.ylabel('team',fontsize=15) 
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(10,20))
plt.barh(tokyo['team'],tokyo['gold']) 
plt.title('teamvice gold medal',fontsize=15,color='red') 
plt.xlabel('gold',fontsize=15) 
plt.ylabel('team',fontsize=15) 
plt.grid(False)
plt.show()
```

###   countplot

```python
plt.figure(figsize=(10,15))
sns.countplot(y='gold',data=tokyo,palette='tab10')
plt.show()
```

###  grouping gold and silver medals  won  by the teams
we can easily visualize the gold and silver medals won by the teams here.gold was noted by the color violet and silver was noted by the colour red.and in this we can see that most gold and silver medals won by the team USA

```python

tokyo.groupby('team')[['gold','silver']].sum().plot.bar(color=['blue','red'],figsize=(15,12))
plt.ylabel('gold/silver')
plt.show()
```

### stripplot:
here the teams who won gold medals , in this we can see that USA  wins most number of gold medals.

```python
plt.figure(figsize=(10,15))
sns.set_style('darkgrid')
sns.stripplot(x='gold',y='team',data=tokyo)

plt.show()
```

### lineplot:
in this we can see that teams who won the most silver medals,here also USA  leads with most silver medals  

```python
plt.figure(figsize=(10,15))
sns.lineplot(x='silver',y='team',data=tokyo)
plt.show()
```

### boxplot:
in this we can visualize that the team who won most number of gold medals,USA have most number of bronze medal in this dataset. 

```python
plt.figure(figsize=(10,15))
sns.boxplot(x='bronze',y='team',data=tokyo,width=5)
plt.show()
```

###  import the library to train ,test and split

### perform the split

###  specify the split size manually by specifying the parameter of the test size and random number generator

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(np.mean(x_train))
print(np.mean(x_test))
print(np.mean(y_train))
print(np.mean(y_test))
```

### linear regression analysis

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x_train,y_train)
print('train test complete')
```

```python
y_pred=l.predict(x_test)
```

```python
tokyo=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
tokyo
```

```python
from sklearn.metrics import mean_absolute_error
print('Mean Absolute Error',mean_absolute_error(y_test,y_pred))

```

```python
sns.regplot(x=y_test,y=y_pred)
plt.title('Regession Plot',size=15)
plt.ylabel('predicted',size=12)
plt.xlabel('actual',size=12)
plt.show()
```
