```python
m=r.groupby('ObservationDate').Deaths.sum().sort_values(ascending=False).head(25)
m
```

```python
n=pd.DataFrame(m)
n
```

```python
n.reset_index(level=0,inplace=True)
n
```

## bar graph of days vs deaths

```python
plt.figure(figsize=(9,5))
ax=plt.axes()
ax.set_facecolor('black')

plt.bar(n.ObservationDate,n.Deaths,label='total deaths',color='g',width=0.5)
plt.legend()
plt.xlabel('days')
plt.ylabel('deaths')
plt.title('most death days')
plt.xticks(rotation=90)
plt.show()
```

## brazilian cases

```python
h=r[r['Region']=='Brazil']
h
```

```python
i=h.groupby('State').Deaths.sum().sort_values(ascending=False).head(10)
i
```

```python
k=pd.DataFrame(i)
k
```

```python
k.reset_index(level=0,inplace=True)
k
```

## pie chart of deaths in brazil

```python
plt.figure(figsize=(9,9))
y=k.Deaths

l=k.State
c=['r','b','g','k','y','r','b','g','k','y']

plt.pie(y,labels=l,startangle=90,shadow=True,colors=c,explode=(0,0,0,0.2,0,0,0,0,0,0))
plt.legend(loc='lower left')
plt.title('deaths in brazilian states')
plt.show()
```

## barplot of brazilian deaths

```python
plt.figure(figsize=(9,9))
plt.title('brazil states death rate')
sns.barplot(x='State',y='Deaths',data=k,palette='coolwarm_r')
plt.xlabel('states',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```
